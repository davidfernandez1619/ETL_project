```python
movies = pd.read_csv('/Users/davidusuariocasa/Desktop/oscars.csv')

movies
```

```python
filtered_df = movies[movies['winner'] == True] # I filter by 'True' only to show the winners and discard the nominated who didn't win (those with value 'False')

filtered_df
```

```python
null_rows = filtered_df[movies.isnull().any(axis=1)] # I do this to see which rows contain null values

no_nulls = filtered_df.dropna()

no_nulls # now I have all the winners, without null values
```

```python
columns_to_drop = ['winner', 'name']

no_nulls.drop(columns=columns_to_drop, inplace=True) # Now I remove the column 'winner', since now I only have winners in my dataframe

```

```python
no_nulls
```

```python
selected_categories = ['OUTSTANDING PICTURE', 'UNIQUE AND ARTISTIC PICTURE', 'OUTSTANDING PRODUCTION', 'SHORT SUBJECT (Cartoon)', 'SHORT SUBJECT (Comedy)', 'SHORT SUBJECT (Novelty)', 'ASSISTANT DIRECTOR', 'MUSIC (Scoring)', 'MUSIC (Song)', 'DANCE DIRECTION', 'SHORT SUBJECT (Color)', 'SHORT SUBJECT (One-reel)', 'SHORT SUBJECT (Two-reel)', 'SPECIAL EFFECTS', 'DOCUMENTARY (Short Subject)', 'DOCUMENTARY (Feature)', 'FOREIGN LANGUAGE FILM', 'BEST PICTURE', 'SHORT FILM (Animated)', 'SHORT FILM (Live Action)', 'SHORT FILM (Dramatic Live Action)', 'ANIMATED FEATURE FILM', 'INTERNATIONAL FEATURE FILM', 'DOCUMENTARY FEATURE FILM', 'DOCUMENTARY SHORT FILM']

finaldf = no_nulls[no_nulls['category'].isin(selected_categories)] # I select only those winning categories that I consider represent best the film itself

finaldf.reset_index(inplace = True) # I reset the indexes
```

```python
finaldf = finaldf.drop(columns='index') # We remove the column 'index'
```

```python
finaldf
```

```python
# Now to do our web scraping, we need to import all these libraries

import numpy as np 
from selenium import webdriver # webdriver permite manejar un navegador 
from webdriver_manager.chrome import ChromeDriverManager # permite instalar y mantener actualizado el driver de chrome
from selenium.webdriver.common.keys import Keys # permite simular teclas del teclado 
from selenium.webdriver.chrome.options import Options # permite configurar el driver de chrome como modo incognito o maximizar la ventana
import time # esperas entre ejecuciones de codigo
import random


import warnings # permite ignorar los warnings de python 
warnings.filterwarnings('ignore')
```

```python
opciones= Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
#para ocultarme como robot
opciones.add_experimental_option('useAutomationExtension', False)
opciones.add_argument('--start-maximized') #empezar maximizado
opciones.add_argument('user.data-dir=selenium') #guarda las cookies
opciones.add_argument('--incognito')#incognito window
```

```python
driver = webdriver.Firefox()

tiempo_espera = random.uniform(0.5, 1)

for index, (value, year) in enumerate(zip(finaldf['film'], finaldf['year_film'])):
    driver = webdriver.Firefox()
    driver.get("https://www.google.com/webhp?hl=es&sa=X&ved=0ahUKEwib6NqMzP7_AhUbVqQEHU22BKcQPAgJ")
    driver.implicitly_wait(2)

    try:
        driver.find_element("css selector","#L2AGLb > div:nth-child(1)").click() # We click to accept the cookies
    except:
        pass

    driver.find_element("css selector","#APjFqb").send_keys(f'"{value} {str(year)}"' + ' ' + 'site:https://www.boxofficemojo.com/', Keys.ENTER) # We select the search box with each film and year

    try:
        driver.find_element("css selector",".LC20lb").click() # If we get results, we click on the link

        try:
            box_office = driver.find_element('css selector','.mojo-performance-summary-table > div:nth-child(4) > span:nth-child(3) > span:nth-child(1)').text # If the page shows the box office, we store the value in a variable called 'box_office'
            finaldf.loc[index, 'box_office'] = box_office

        except:
            finaldf.loc[index, 'box_office'] = np.nan # If not, we store the value as null
    except:
        finaldf.loc[index, 'box_office'] = np.nan

    driver.quit()
    time.sleep(2)  # Esperar 2 segundos antes de abrir el navegador nuevamente

driver.quit()
```
