```python
import pandas as pd

from collections import Counter
from fuzzywuzzy import fuzz
import pickle
from time import sleep
```

```python
movies = pd.read_csv('/Users/davidusuariocasa/Desktop/IMDb_movies.csv')

movies
```

```python
sorted_df = movies.sort_values('year') # We order the years included in the dataframe

sorted_df['year'].unique() # We see the numbers of sorted years included in the dataframe
 
```

```python
USA = sorted_df[sorted_df['country'] == 'USA'] # We only want the movies made in 'USA'
```

```python
# We have a lot of genres. We want to assign each movie one of the top 10 most frequent genres

USA['genre'] = USA['genre'].str.lower() # We first set every genre in lowercase just in case
```

```python
pd.DataFrame(USA['genre'].value_counts()/len(USA)*100).head(40) # We see, for each genre, what's it's percentage
```

```python
cuenta_generos = Counter(genres for genres in USA['genre']) # This shows me how many total genres there is in my dataframe
print(f"Hay {len(cuenta_generos)} generos diferentes.") 
```

```python
comunes = dict(cuenta_generos.most_common(10)) # I create the top 10 most common genres based on the previous dataframe
comunes
```

```python
def generos_(col, generos_comunes):  # We create a function that categorizes each genre among the top 10 most common genres based on its similarity with them
    maximo = 0
    for key in generos_comunes.keys():
        parecido = fuzz.ratio(col, key) 
        if parecido > maximo:
            maximo = parecido
            
            genero = key
    
            
    # We make sure that the genres are at least 50% similar        
    if maximo > 50:
        return genero
    else:
        return "Other"
```

```python
USA["genre2"] = USA.apply(lambda x: generos_(x["genre"], comunes), axis = 1) # We create a new column called 'genre2' that classifies each film among the top 10 most common genres
```

```python
# Now we want to keep cleaning our DF, so we remove columns we don't want

columns_to_drop = ['imdb_title_id', 'title', 'date_published', 'genre', 'duration', 'language', 'writer', 'production_company', 'description', 'avg_vote', 'votes', 'budget', 'usa_gross_income', 'worlwide_gross_income', 'metascore', 'reviews_from_users', 'reviews_from_critics']

USA = USA.drop(columns=columns_to_drop) # We create a dictionary with all the columns we want to remove, and we remove then next
```

```python
USA = USA.rename(columns={'genre2': 'genre'})

USA = USA.rename(columns={'original_title': 'film'})

USA.reset_index(inplace = True)

# We change the names of some columns, so that they match the Oscars CSV
```

```python
USA = USA.drop(columns='index') # After our reset_index, we remove the column 'index'
```

```python

USA = USA.reset_index(drop=True) # And we reset our indexes

USA
```
