```python
import pandas as pd
```

```python
movies = pd.read_csv('/Users/davidusuariocasa/Desktop/oscars.csv') # I open the csv with the Oscar winners from 1928 to 2023

movies
```

```python
filtered_df = movies[movies['winner'] == True] # I filter by 'True' only to show the winners and discard the nominated who didn't win (those with value 'False')

filtered_df
```

```python
null_rows = filtered_df[movies.isnull().any(axis=1)] # I do this to see which rows contain null values

finaldf = filtered_df.dropna() # And now I drop all my null values

finaldf # now I have all the winners, without null values
```

```python
finaldf.reset_index(inplace = True) # Now I reset all indexes in my df

finaldf
```

```python
columns_to_drop = ['winner', 'index'] # I create a list with the names of the columns I want to remove from my df

finaldf.drop(columns=columns_to_drop, inplace=True) # Now I remove the column 'winner' and 'index'

```

```python
finaldf # This is my final df
```

```python
finaldf.to_csv('Oscars_cleaned.csv', index=False)
```
